```python
from astropy.io import ascii
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def plot_data(ascii_tbl, label_str = None, input_colour='#FF9933'):
    plt.plot(ascii_tbl['col1'],ascii_tbl['col2'], label = label_str, color= input_colour)
```

```python
# import ukidds data
UKIDDS_H = ascii.read('data/UKIRT-UKIDSS.H.dat')
UKIDDS_K = ascii.read('data/UKIRT-UKIDSS.K.dat')
UKIDDS_Z = ascii.read('data/UKIRT-UKIDSS.Z.dat')
UKIDDS_J = ascii.read('data/UKIRT-UKIDSS.J.dat')
UKIDDS_Y = ascii.read('data/UKIRT-UKIDSS.Y.dat')

# import galex data
GALEX_FUV = ascii.read('data/GALEX-GALEX.FUV.dat')
GALEX_NUV = ascii.read('data/GALEX-GALEX.NUV.dat')

# import sloan data
SLOAN_u = ascii.read('data/SLOAN-SDSS.u.dat')
SLOAN_g = ascii.read('data/SLOAN-SDSS.g.dat')
SLOAN_r = ascii.read('data/SLOAN-SDSS.r.dat')
SLOAN_i = ascii.read('data/SLOAN-SDSS.i.dat')
SLOAN_z = ascii.read('data/SLOAN-SDSS.z.dat')

# import wise data
WISE_W1 = ascii.read('data/WISE-WISE.W1.dat')
WISE_W2 = ascii.read('data/WISE-WISE.W2.dat')
WISE_W3 = ascii.read('data/WISE-WISE.W3.dat')
WISE_W4 = ascii.read('data/WISE-WISE.W4.dat')
```

```python
plt.figure(figsize=(20,10))

# plot galex
plot_data(GALEX_FUV, 'GALEX', 'b')
plot_data(GALEX_NUV, input_colour='b')
plt.text(GALEX_NUV['col1'][0]-100,0.65, 'GALEX', fontsize=20)

# plot sloan
plot_data(SLOAN_u, 'SDSS', 'g')
plot_data(SLOAN_g, input_colour='g')
plot_data(SLOAN_r, input_colour='g')
plot_data(SLOAN_i, input_colour='g')
plot_data(SLOAN_z, input_colour='g')
plt.text(SLOAN_r['col1'][0],0.65, 'SDSS', fontsize=20)

# plot ukidds
plot_data(UKIDDS_Z, 'UKIDDS')
plot_data(UKIDDS_Y)
plot_data(UKIDDS_J)
plot_data(UKIDDS_H)
plot_data(UKIDDS_K)
plt.text(15000,0.65, 'UKIDDS', fontsize=20)

plt.ylabel('Transmission',fontsize=18)
plt.xlabel('Wavelength $\AA$',fontsize=18)
plt.xticks(fontsize=10)

plt.title('Filter profile for the multiwavelength catalogue excluding WISE')



plt.savefig('filter_profiles.png')
```
