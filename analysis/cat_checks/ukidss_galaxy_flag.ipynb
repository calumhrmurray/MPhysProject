```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def plot_hist(array,bin_num,label,default=True):
    n, bins = np.histogram(array,bins=bin_num,normed=default)
    plt.plot(bins[1:],n,label=label)
    return 

def colour_plot_hist(array,label,bin_num=1000,default=False):
    array = [row[0]-row[1] for row in array]
    n, bins = np.histogram(array,bins=bin_num,normed=default)
    plt.plot(bins[1:],n,label=label)
    return 

def petro_plot_hist(array,label,bin_num=1000,default=False):
    array = [row[1] for row in array]
    n, bins = np.histogram(array,bins=bin_num,normed=default)
    plt.plot(bins[1:],n,label=label)
    return 

def surface_brightness_plot_hist(array,label,bin_num=1000,default=False):
    array = [row[1]+2.5*np.log10(2*np.pi*row[4]*row[4]) for row in array]
    n, bins = np.histogram(array,bins=bin_num,normed=default)
    plt.plot(bins[1:],n,label=label)
    return 

# 
# each array has the exact same rows: 0 u, 1 r, 2 mr_petro, 3 z, 4 petroR50_r
```

```python
unflagged = np.load('/home/calum/Documents/Mphys_data/cat_check/ukidss_allbands.npy')
flagged = np.load('/home/calum/Documents/Mphys_data/cat_check/ukidss_notallbands.npy')

len(unflagged), len(flagged)
```

```python
plt.figure(figsize=(20,10))
colour_plot_hist(unflagged,'unflagged',default=True)
colour_plot_hist(flagged,'flagged',default=True)
plt.xlim(0,5)
plt.legend()
```

```python
plt.figure(figsize=(20,10))
petro_plot_hist(unflagged,'unflagged',default=True)
petro_plot_hist(flagged,'flagged',default=True)
plt.legend()
```

```python
plt.figure(figsize=(20,10))
surface_brightness_plot_hist(unflagged,'unflagged',default=True)
surface_brightness_plot_hist(flagged,'flagged',default=True)
plt.legend()
```
